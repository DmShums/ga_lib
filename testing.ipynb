```python
import random_graph
import subprocess
import statistics
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_distance_gen_result = [15.0, 23.8, 50.1, 71.8]
total_distance_classic_result = [15, 23, 45, 53]
generations = ['5 nodes', '10 nodes', '15 nodes', '20 nodes']
```

```python
plt.figure(figsize=(10, 6))

std_dev_gen = []
# std_dev_classic = np.std(total_distance_classic_result, ddof=1)

for i in range(4):
    std_dev_gen.append(np.std([total_distance_gen_result[i], total_distance_classic_result[i]], ddof=1))

std_dev_gen_points = std_dev_gen

print(std_dev_gen_points)
# std_dev_classic_points = [std_dev_classic] * len(generations)

bar_width = 0.35

r1 = np.arange(len(total_distance_gen_result))
r2 = [x + bar_width for x in r1]

plt.bar(r1, total_distance_gen_result, color='blue', width=bar_width, yerr=std_dev_gen_points, capsize=5, label='Gen Algorithm')
plt.bar(r2, total_distance_classic_result, color='red', width=bar_width, capsize=5, label='Classic Method')

plt.xlabel('Size of Graph')
plt.ylabel('Total Distance')
plt.xticks([r + bar_width/2 for r in range(len(total_distance_gen_result))], generations)

plt.title('Total Distance Comparison with Standard Deviation')
plt.legend()

plt.grid(True)
plt.show()
```
